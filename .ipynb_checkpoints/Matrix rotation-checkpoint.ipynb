```c++
#import <vector>
#import <random>

namespace amazon_interview {
    
    using namespace std;
    
    vector<vector<int> > generate_matrix(int dimension) {
        vector<vector<int> > matrix;
        
        int rows = dimension;
        
        while (rows > 0) {
            
            int cols = dimension;
            
            vector<int> column;
            
            while (cols > 0) {
                
                int v = (rand()%19) + (-9);
                
                column.push_back(v);
                
                cols--;
            }
            
            matrix.push_back(column);
            
            rows--;
        }
        
        return matrix;
    }
    
    void print_matrix(string header, vector<vector<int> > & matrix) {
        
        cout << header << "\n";
        
        for (vector<int> row : matrix) {
            
            for (int i : row) {
                cout << setw(2) << i << ", ";
            } 
            
            cout << "\n";
        }
        
        cout << "\n";
    }
    
    void swap_elements(int & e1, int & e2) {
        int tmp = e1;
        e1 = e2;
        e2 = tmp;
    }
    
    void swap_row_to_column(vector<vector<int> > & m, int layer, int length, int start_col) {
      
        //cout << "Layer: " << layer << " length: " << length << " start_col: " << start_col << "\n\n";
        int max_col = length + layer - 1;
        int col1 = layer;
        int row = layer;
        int new_col = max_col;
        int offset = 0;
        
        if (new_col == start_col) {
            row = new_col;
            new_col = layer;
            offset = 1; 
        }
        
        while (col1 < max_col) {
            
            int col = col1 + offset;
            int new_row = col;
            swap_elements(m[row][col], m[new_row][new_col]);
            //cout << "From row:" << row << " col:" << col << " to row:" << new_row << " col:" << new_col << "\n";
            col1++;
        } 
    }
    
    void swap_columns(vector<vector<int> > & m, int layer, int length) {
        int max_row =  length + layer - 1;
        int new_row = max_row;
        int new_col = new_row;        
        int col = layer;
        int row = layer;        
        
        //cout << "Layer: " << layer << " length: " << length << "\n";
        
        while (col < max_row) {
            
            swap_elements(m[row][col], m[new_row][new_col]);
            //cout << "From row:" << row << " col:" << col << " to row:" << new_row << " col:" << new_col << "\n";
            col++;
            new_col--;
        }
    }
    
    void rotate_matrix(vector<vector<int> > & m) {
        
        int matrix_size = m.size();
        
        int layer = 0;
        
        int layer_size = matrix_size - layer * 2;        
        
        while (layer_size >= 2) {  
            
            swap_row_to_column(m, layer, layer_size, layer);
            
            swap_row_to_column(m, layer, layer_size, (layer_size + layer - 1));
            
            swap_columns(m, layer, layer_size);
            
            layer++;
            
            layer_size = matrix_size - layer * 2;
        }
    }

}
```

```c++
auto matrix = amazon_interview::generate_matrix(11);

amazon_interview::print_matrix("Matrix", matrix);

amazon_interview::rotate_matrix(matrix);

amazon_interview::print_matrix("Rotated Matrix", matrix);
```
